```python
np.random.seed(50)
a = np.random.randint(10, 20, (3, 4))
b = np.random.randint(30, 40, (2, 6))
a, b
```

```python
b.shape = (3, 4)  # перед тем как поэлементно складывать или умножать массивы, нужно их привести к одной размерности 
```

```python
a + b
```

```python
a*b
```

```python
b.shape = (4, 3) # при матричном умножении "внутренние" размерности должны совпадать
b
```

```python
res = a @ b # получили массив с размерностью 3 на 3
res
```

```python
res.mean() # среднее арифмитическое всех элементов
```

```python
res.max() # максимальный из всех элементов массива
```

```python
res[2].mean() # среднее арифмитическое элементов строки с индексом 2
```

```python
res[:, 2].mean() # среднее арифмитическое элементов столбца с индексом 2
```
